```python
from datetime import datetime, timezone
from data_decay import FinancialFreshnessScorer

freshness_scorer = FinancialFreshnessScorer()
current_time = datetime.now((timezone.utc))

for article in articles_ret:
    freshness = freshness_scorer.calculate_confidence(
        article,
        current_time
    )
    # print(freshness)
    boost = freshness_scorer.get_boost_factor(freshness['confidence_score'])
    article.score = article.score * boost  # Adjust the score
    print("The Score Becomes: ", article.score, "Article ID:", article.article_id)
```

```python
from pipeline.embeddings import EmbeddingModelSingleton, CrossEncoderModelSingleton
from pipeline.qdrant import build_qdrant_client
from pipeline.retriever import QdrantVectorDBRetriever, RetrievalVisualizer
```

```python
embedding_model = EmbeddingModelSingleton()
cross_endocer_model = CrossEncoderModelSingleton()
qdrant_client = build_qdrant_client()

vector_db_retriever = QdrantVectorDBRetriever(
    embedding_model=embedding_model,
    vector_db_client=qdrant_client,
    cross_encoder_model=cross_endocer_model,
)
```

```python
all_posts = vector_db_retriever.scroll()
visualizer = RetrievalVisualizer(posts=all_posts)
```

```python
query_question = "Posts about Qdrant"
retrieved_results = vector_db_retriever.search(query=query_question, limit=3, return_all=True)
for post in retrieved_results["posts"]:
    vector_db_retriever.render_as_html(post)
```

```python
visualizer.render(
    embedded_queries=retrieved_results["embdedded_queries"],
    retrieved_posts=retrieved_results["posts"],
)
```
